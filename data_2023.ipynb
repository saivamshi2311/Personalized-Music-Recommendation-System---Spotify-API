```python
data_2023.drop_duplicates(inplace=True)
```

```python
print(data_2023['streams'].dtype)
```

```python
data_2023['streams'] = pd.to_numeric(data_2023['streams'], errors='coerce')
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Example: Box plot for 'streams' 
plt.figure(figsize=(8, 6))
sns.boxplot(x=data_2023['streams'], color='green')  # Changing the color to sky blue
plt.title('Boxplot for Streams')
plt.show()

```

```python

# Correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(data_2023.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Pairplot for selected features
sns.pairplot(data_2023[['danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'streams']])
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()

# Boxplot for key vs. danceability
plt.figure(figsize=(12, 6))
sns.boxplot(x='key', y='danceability_%', data=data_2023)
plt.title('Boxplot of Danceability Across Keys')
plt.show()

# Distribution of BPM
plt.figure(figsize=(10, 6))
sns.histplot(data_2023['bpm'], bins=30, kde=True, color='orange')  # Changing the color to orange
plt.title('Distribution of BPM')
plt.show()

# Bar plot for artist count
plt.figure(figsize=(12, 6))
sns.countplot(x='artist_count', data=data_2023)
plt.title('Distribution of Artist Count')
plt.show()

# Scatter plot for energy vs. valence
plt.figure(figsize=(10, 6))
sns.scatterplot(x='energy_%', y='valence_%', data=data_2023, hue='mode', palette='viridis', size='streams', sizes=(20, 200))
plt.title('Scatter Plot of Energy vs. Valence')
plt.show()
```
