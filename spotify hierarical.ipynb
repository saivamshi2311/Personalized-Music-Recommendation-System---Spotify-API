```r
# Assuming your dataframe is named 'df' and you want to drop columns named 'column1' and 'column2'
numeric_data <- subset(data, select = -c(track_id, artists, album_name, track_name, explicit))
```

```r
numeric_data
```

```r
numeric_data
```

```r
# Scale the numeric data
scaled_data <- scale(numeric_data)

# Check for missing values
if (anyNA(scaled_data)) {
  stop("Scaled data contains missing values. Please handle missing values before proceeding.")
}

# Calculate cosine similarity
cosine_sim <- tcrossprod(scaled_data) / (sqrt(rowSums(scaled_data^2)) %*% t(sqrt(rowSums(scaled_data^2))))

# Perform hierarchical clustering using cosine similarity
hc_cosine <- hclust(as.dist(1 - cosine_sim))

# Increase plot size
options(repr.plot.width = 50, repr.plot.height = 25) # Adjust width and height as needed

# Plot the dendrogram
plot(hc_cosine, main = "Hierarchical Clustering Dendrogram (Cosine Similarity)", xlab = "Samples", sub = NULL, cex = 0.6)
```

```r
# Increase plot size and font size for better readability
options(repr.plot.width = 12, repr.plot.height = 8) # Adjust width and height as needed
par(cex = 1.2)  # Adjust font size as needed

# Plot the dendrogram with improved appearance
plot(hc_cosine, main = "Hierarchical Clustering Dendrogram (Cosine Similarity)", xlab = "Samples", sub = NULL, cex = 0.6)

# Add labels with larger font size
labels <- as.character(rownames(scaled_data))
labels <- strwrap(labels, width = 10)  # Adjust width as needed for better text wrapping
labels <- sapply(labels, paste, collapse = "\n")
labels <- gsub(" ", "\n", labels)

text(hc_cosine, labels = labels, hang = -1, cex = 1.2)  # Adjust cex for label font size
```
